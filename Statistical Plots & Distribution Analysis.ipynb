```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("customer_data.csv")

# Display dataset
print(df.head())

# -------------------------------
# Histogram with KDE (Sales)
# -------------------------------
plt.figure(figsize=(7,5))
sns.histplot(df["Sales"], bins=10, kde=True, color="skyblue")
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.savefig("sales_histogram.png")
plt.show()

# -------------------------------
# KDE Plot (Age)
# -------------------------------
plt.figure(figsize=(7,5))
sns.kdeplot(df["Age"], fill=True, color="green")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.savefig("age_kde.png")
plt.show()

# -------------------------------
# Boxplot (Sales)
# -------------------------------
plt.figure(figsize=(7,5))
sns.boxplot(x=df["Sales"], color="orange")
plt.title("Sales Boxplot")
plt.savefig("sales_boxplot.png")
plt.show()

# -------------------------------
# Compare Sales by Region
# -------------------------------
plt.figure(figsize=(7,5))
sns.boxplot(x="Region", y="Sales", data=df)
plt.title("Sales Distribution by Region")
plt.savefig("sales_region_boxplot.png")
plt.show()

# -------------------------------
# Compare Age by Region
# -------------------------------
plt.figure(figsize=(7,5))
sns.histplot(data=df, x="Age", hue="Region", kde=True)
plt.title("Age Distribution by Region")
plt.savefig("age_region_histogram.png")
plt.show()

# -------------------------------
# Outlier Detection (Sales)
# -------------------------------
Q1 = df["Sales"].quantile(0.25)
Q3 = df["Sales"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["Sales"] < lower) | (df["Sales"] > upper)]

print("\nOutliers:")
print(outliers)

# -------------------------------
# Skewness & Spread
# -------------------------------
print("\nSales Skewness:", df["Sales"].skew())
print("Sales Standard Deviation:", df["Sales"].std())

# -------------------------------
# Interpretation
# -------------------------------
print("""
Interpretation:
The histogram and KDE plot show the overall distribution of Age and Sales.
The boxplot highlights any outliers in the Sales column.
The regional comparison shows differences in sales across Regions A, B, C, and D.
The skewness value indicates whether the Sales distribution is symmetric or skewed.
The standard deviation measures the spread of Sales values.
All plots have been saved as PNG files.
""")
```
